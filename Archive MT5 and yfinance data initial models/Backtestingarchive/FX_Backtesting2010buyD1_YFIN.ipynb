```python
import yfinance as yf
from backtesting import Backtest, Strategy
import pandas as pd
from datetime import datetime

symbol = "USDCAD=X"
ohlc_data = yf.download(symbol,start="2022-10-25", end=datetime.now(), progress=False)


# Select only the required columns
df = ohlc_data[['Open', 'High', 'Low', 'Close']]

df_pred = pd.read_csv('prediction2010buy90.csv', index_col=0)

class MyStrategy(Strategy):
    def init(self):
        self.count = 0

    def next(self):
        current_time = self.data.index[-1]
        current_time_str = current_time.strftime('%Y-%m-%d %H:%M:%S')
 
        if current_time_str in df_pred.index:
            prediction = df_pred.loc[current_time_str, 'prediction']
            if prediction == 1:
                self.count += 1
                price = self.data.Open[-1]
                self.buy(size=100000, sl=price - 0.010, tp=price + 0.030)

# Create and run backtest
bt = Backtest(df, MyStrategy, cash=10000, commission=.0003, margin = 0.01)
output = bt.run()
print(output)

# Plot results
#bt.plot()
```
